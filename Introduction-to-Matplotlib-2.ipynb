### Which one to use (Pyplot VS Matplotlib oo method) ?

* When Plotting Something Quickly okay to use pyplot method
* * When plotting something more advanced use the oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease.age>50]
over_50
```

```python
# Pyplot method
over_50.plot(kind="scatter",x='age',y='chol',c="target")
plt.show()
```

```python
# OO method
fig ,ax = plt.subplots(figsize=(10,6))
ax.scatter(over_50.age,over_50.chol)
plt.show()
```

```python
# Oo method mixed with pyplot method
fig , ax = plt.subplots()
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,100])
#Sets Minimum and maxmum x limit
plt.show()
```

```python
## OO Method from Scratch
fig ,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(over_50.age,over_50.chol,c=over_50.target)
ax.set(title="Heart Diasease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol Level")


# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizintal Line
ax.axhline(over_50.chol.mean(),ls="--")
# ls stands for line style no line style then complete line

plt.show()
```

```python
over_50.head()
```

## Subplot of chol, age ,thalach

```python
fig , (ax0 ,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter = ax0.scatter(over_50.age,over_50.chol,c=over_50.target)
ax0.set(title="Heart disease and Cholestrol Levels",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50.chol.mean(),ls="--")

scatter2 =ax1.scatter(over_50.age,over_50.thalach,c=over_50.target)
ax1.set(title="Age and Max Heart Rate",xlabel="Age",ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(),title="target")
ax1.axhline(over_50.thalach.mean(),ls="--")

#Add a title to the complete figure
fig.suptitle("Heart Disease Analysis")


plt.show()
```

```python
# How to change style from Pre Existiong Style

plt.style.use("seaborn-v0_8-whitegrid")

## OO Method from Scratch
fig ,ax = plt.subplots(figsize=(10,6))

#cmap="winter"
# For More colors visit Documentation
# we can also use plasma or summer
scatter = ax.scatter(over_50.age,over_50.chol,c=over_50.target, cmap="winter")
ax.set(title="Heart Diasease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol Level")


# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizintal Line
ax.axhline(over_50.chol.mean(),ls="--",c="red")
# ls stands for line style no line style then complete line

plt.show()
```

```python
# Customising y and x axis limitaion

fig , (ax0 ,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter = ax0.scatter(over_50.age,over_50.chol,c=over_50.target,cmap="winter")
ax0.set(title="Heart disease and Cholestrol Levels",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50.chol.mean(),ls="--")
#Change the x axis limit
ax0.set_xlim([50,80])
ax0.set_ylim([100,430])

scatter2 =ax1.scatter(over_50.age,over_50.thalach,c=over_50.target,cmap="winter")
ax1.set(title="Age and Max Heart Rate",xlabel="Age",ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(),title="target")
ax1.axhline(over_50.thalach.mean(),ls="--")

#change ax1 x axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#Add a title to the complete figure
fig.suptitle("Heart Disease Analysis")


plt.show()
```
