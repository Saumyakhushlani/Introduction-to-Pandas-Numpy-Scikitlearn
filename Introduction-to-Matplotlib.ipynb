## Matplotlib Workflow Example

```python
x = [1,2,3,4]
y = [11,22,33,44]
```

#### 2. Setup Plot (Object Oriented)

```python
fig , ax = plt.subplots(figsize=(10,10))

# 3. Plot/Add Data
ax.plot(x,y)
```

## Making Figures with Numpy Arrays

We want:
* line plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the Data and Create a line plot
#line Plot
fig , ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# Using same data to make a Scatter Plot
#scatter Plot
fig , ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# Another Scatter Plot
fig , ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# Make a plot from Dictionary
# Bar Plot
nut_butter_prices = {"Almond butter":10,
                    "Peanut Butter":8,
                    "Cashew Butter":12}
fig , ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter Prices",xlabel="Type of butter",ylabel="Price")
```

```python
#Horizontal Bar graph
fig ,ax = plt.subplots()
ax.barh(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter Prices",xlabel="Type of butter",ylabel="Price")
```

```python
# Histogram
#make some data
#randn return a sample (or samples) from the "standard normal" distribution.
x= np.random.randn(1000)
fig ,ax = plt.subplots()
ax.hist(x);
```

### Subplots (Multiple Plots Happening at the same time)

> Two options for Subplot

```python
#option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# Plot data to each diffrerent Axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# Option 2
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(100),np.random.random(100))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

## Ploting from Pandas Dataframe

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2025",periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,.]', '', regex=True)
car_sales
```

```python
# Remove Last 2 Zeroes
car_sales.Price = car_sales.Price.str[:-2]
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2025",periods=len(car_sales))
car_sales
```

```python
car_sales["Total"] = car_sales.Price.astype(int).cumsum()
# astype Converts string to int temporary and cumsum add cumilative sum
car_sales
```

```python
# Lets plot the total Sale
car_sales.plot(x="Sales Date",y="Total")
```

```python
car_sales.plot(x="Odometer (KM)",y="Total",kind="scatter")
```
