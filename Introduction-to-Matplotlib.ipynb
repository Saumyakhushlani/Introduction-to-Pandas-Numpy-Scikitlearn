## Matplotlib Workflow Example

#### 0. Import matplotlib and get it ready for plotting in jupyter Notebook

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

#### 1. Prepare Data

```python
x = [1,2,3,4]
y = [11,22,33,44]
```

#### 2. Setup Plot (Object Oriented)

```python
fig , ax = plt.subplots(figsize=(10,10))

# 3. Plot/Add Data
ax.plot(x,y)
```

#### 4. Customise Plot

```python
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
```

#### 5. Save and Show data

```python
fig.savefig("images/sample-plot.png")
plt.show()
```
