# Manuplating Data

```python
#We can acess the string Value of Column by .str
car_sales.Make.str.lower()
#But it will not change the DataFrame because we don't reassign it.
```

```python
car_sales.Make = car_sales.Make.str.lower()
car_sales
```

```python
#All the times the data may not be filled
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
#To fill the missing data with mean of the other data
car_sales_missing.Odometer.fillna(car_sales_missing.Odometer.mean())
```

```python
car_sales_missing
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
#To remove na values
#inplace=True means replace the values in the table
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

## Creating new Column

```python
# Column from Series
seats_colums =pd.Series([5,5,5,5,5])

#Adding New Column
car_sales["Seats"]= seats_colums
car_sales
```

```python
car_sales.Seats.fillna(5,inplace=True)
car_sales
```

```python
# Column from Pythin List
Fuel_economy = [7.5,9.2,5.0,9.6,8.7,9.2,7.8,8.4,6.9,8.8]
car_sales["Milage"] = Fuel_economy
car_sales
```

```python
# Column from Exisitng Column
car_sales["Total fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Milage"]
car_sales
```
